각 테스터들은, 각각 몇개의 와인을 테스트 했는지 확인하시오. ( 테스터 이름, 갯수 )

```python
#### 나의 풀이


```

```python
#### 해설

# 각 테이스터 별로 몇개의 데이터가 있느냐 라는 의미


# 1. 첫번째 방법 - groupby
# ~~ 별로 묶을 때
# 테이스터 네임별로 묶어달라
reviews_ti.groupby('taster_name')
# 묶은 후 관심있는 컬럼 적기
reviews_ti.groupby('taster_name')['taster_name']
# 묶은 후 관심있는 컬럼 의 숫자 세기
reviews_ti.groupby('taster_name')['taster_name'].count()

# 위 데이터는 시리즈
# 고로 시리즈 함수 사용 가능 : 예) sum()

```

```python
#### 해설

# 2. 두번째 방법 - .value_counts

reviews_ti['taster_name'].value_counts()
```

리뷰의 포인트의 평균을 구하고, 리뷰의 포인트값이, 평균보다 큰 데이터 (즉, 평가가 좋은 와인) 만 가져오시오.

```python
#### 나의 풀이

# 1. 리뷰포인트 평균 구하는 코드 가져오기
reviews['points'].mean()
# 2. 평균보다 높은 데이터만 가져오기
reviews[(reviews['points']) >= (reviews['points'].mean())]
# 3. 중복데이터 제거하고 와인 이름만 남기기
```

```python
#### 해설 


# 1. 포인트 칼럼 평균 구하기
reviews_ti['points'].mean()

# 2. 평균보다 높은 행 구하기
reviews_ti['points'] > reviews_ti['points'].mean()

# 3. True/ False 중 True 인 행 가져오기
reviews_ti.loc[ reviews_ti['points'] > reviews_ti['points'].mean() ,  ]


```
